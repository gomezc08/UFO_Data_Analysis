```python
# Importing necessary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing dataframe

df = pd.read_csv('/home/ramseyc/Downloads/ufo_data2.csv')
```

```python
# Checking to make sure everything came in correctly

df.head()
```

```python
# Adding a new column for the place

df['Place'] = df['City'] + ', ' + df['State']
```

```python
df.head()
```

```python
# Making the dataframe only have 

df = df[df['Country'] == 'USA']
df.head()
```

```python
# Importing a new type of plot to try

from joypy import joyplot
```

```python
# Creating a dictionary of the months and their names

monthDict = {1.0: 'January',
             2.0: 'February',
             3.0: 'March',
             4.0: 'April',
             5.0: 'May',
             6.0: 'June',
             7.0: 'July',
             8.0: 'August',
             9.0: 'September',
             10.0: 'October',
             11.0: 'November',
             12.0: 'December'
            }
```

```python
# Creating a new column of the string version of the month

df['MonthWord'] = df['Month'].replace(monthDict)
df.head()
```

```python
# With help from https://towardsdatascience.com/ridgeline-plots-the-perfect-way-to-visualize-data-distributions-with-python-de99a5493052

# Using a ridgeline plot to show the density of sightings throughout the month

from matplotlib import cm

plt.figure()

joyplot(
    data = df[['Day', 'Month']], by = 'Month',figsize=(12, 8), colormap = cm.summer_r
)

plt.ylabel('MonthWord', fontsize=16)


unique_months = df['MonthWord'].unique()
plt.yticks(range(len(unique_months)), unique_months)

plt.title('Range of UFO Sightings Across the Month', fontsize=20)
plt.show()
```
